```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('classic')
import matplotlib.font_manager as font_manager
```

```python
plt.rcParams['figure.figsize']=(20,150)
#plt.rcParams.update({'font.size': 15})

title_font = {'fontname':'Arial', 'size':'40', 'color':'black', 'weight':'normal',
  'verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'fontname':'Arial', 'size':'30'}

plt.scatter(sets.year,
            sets.name,
            s=sets.num_parts*5,
            alpha=0.3,
            c=sets.num_parts,
            cmap='viridis'
            )
plt.xlabel('Years', **axis_font)
plt.ylabel('Number of Each Set', **axis_font)
plt.title('Number of Parts by Set by Year',**title_font)
plt.xticks(fontsize='20')
plt.axis([1945, 2025,0,700])
#plt.colorbar()
plt.grid(True)
plt.show()
```
